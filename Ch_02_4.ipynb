**import 套件**

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as plt_font
twfont1=plt_font.FontProperties(fname="drive/My Drive/DL_CLASS/kaiu.ttf")
```

**導入資料**

```python
#利用np.loadtxt由SalesVolume.csv檔案載入資料集
data=np.loadtxt("drive/MyDrive/DL_CLASS/SalesVolume.csv",delimiter=',')
#將第1列資料為廣告推播次數取出為data_x
data_x=data[0]
#將第2列資料為商品銷售數量取出為data_y
data_y=data[1]
print(data_x)
print(data_y)
```

**畫出分佈圖**

```python
plt.figure(figsize=(12,6))
plt.title("產品銷售數量和廣告推播次數",fontproperties=twfont1,fontsize=20)
plt.xlabel("廣告推播次數x",fontproperties=twfont1,fontsize=20)
plt.ylabel("產品銷售數量y",fontproperties=twfont1,fontsize=20)
#利用plt.plot()畫出data_x和data_y的關係圖
plt.scatter(data_x,data_y)
plt.show()
```
